```python
# Fonction pour charger une police de caractères à partir d'une URL
def load_font_from_url(font_url, font_size):
    response = requests.get(font_url)
    if response.status_code == 200:
        font_data = BytesIO(response.content)
        font = ImageFont.truetype(font_data, font_size)
        return font
    else:
        print("Failed to load font from URL.")
        return None

# URL de la police de caractères en ligne
font_url = "https://www.font-police.com/classique/sans-serif/arialbd.ttf"  # Remplacez "https://example.com/font.ttf" par l'URL de votre police de caractères

# Taille de la police de caractères
font_size = 20

# Charger la police de caractères à partir de l'URL
font = load_font_from_url(font_url, font_size)
```

```python
import json

class LinkedinAutomate:
    def __init__(self, access_token, description):
        self.access_token = access_token
        self.description = description
        self.headers = {
            'Authorization': f'Bearer {access_token}',
            'Content-Type': 'application/json'
        }
        self.user_id = None

    @staticmethod
    def load_image_from_url(url):
        response = requests.get(url)
        if response.status_code == 200:
            image = Image.open(BytesIO(response.content))
            return image
        return None

    def generate_images(self, records_data,logo_url,advertisement_file_path):
        try:
          # URL de la police de caractères en ligne
            font_url = "https://www.font-police.com/classique/sans-serif/arialbd.ttf"  # Remplacez "https://example.com/font.ttf" par l'URL de votre police de caractères

            title_font = load_font_from_url(font_url, font_size=35)
            text_font = load_font_from_url(font_url, font_size=30)
            watermark_font = load_font_from_url(font_url, font_size=18)

        except IOError:
            raise Exception("Font file is missing. Ensure the font files are in the correct directory.")

        file_paths = []
        for row in records_data.itertuples():
            width, height = 1080, 1080
            image = Image.new('RGB', (width, height), color='#003C57')
            draw = ImageDraw.Draw(image)

            logo = self.load_image_from_url(logo_url)
            if logo:
                logo = logo.resize((250, 250))
                image.paste(logo, ((width - logo.width) // 2, 10))
            else:
                continue

            title_text = row.Titre
            wrapped_title = textwrap.wrap(title_text, width=60)
            current_h, pad = 110+logo.height, 5
            for line in wrapped_title:
                bbox = draw.textbbox((0, current_h), line, font=title_font)
                w, h = bbox[2] - bbox[0], bbox[3] - bbox[1]
                draw.text(((width - w) // 2, current_h), line, fill='#EBB016', font=title_font)
                current_h += h + pad

            main_text = row.Résumé
            wrapped_text = textwrap.wrap(main_text, width=60)
            current_h += 40
            for line in wrapped_text:
                bbox = draw.textbbox((0, current_h), line, font=text_font)
                w, h = bbox[2] - bbox[0], bbox[3] - bbox[1]
                draw.text(((width - w) // 2, current_h), line, fill='#F4EBD6', font=text_font)
                current_h += h + pad

                 # Obtenir la date d'hier au format approprié
                date_hier = (datetime.now() - timedelta(days=1)).strftime("%Y-%m-%d")

                # Utiliser la date d'hier dans le texte du filigrane
                watermark_text = "Actualité du " + date_hier + " - Source: journal Lemonde"



            bbox = draw.textbbox((width, height), watermark_text, font=watermark_font)
            w, h = bbox[2] - bbox[0], bbox[3] - bbox[1]
            draw.text((width - w - 10, height - h - 10), watermark_text, fill='white', font=watermark_font)

            file_name = f'sharepoint_text_image_{row.Index}.png'
            image.save(file_name)
            file_paths.append(file_name)
        file_paths.append(advertisement_file_path)
        return file_paths

    def register_media(self, image_file_path):
        register_url = "https://api.linkedin.com/v2/assets?action=registerUpload"
        headers = self.headers.copy()
        headers['Content-Type'] = 'application/json'
        data = {
            "registerUploadRequest": {
                "recipes": [
                    "urn:li:digitalmediaRecipe:feedshare-image"
                ],
                "owner": f"urn:li:person:{self.user_id}",
                "serviceRelationships": [
                    {
                        "relationshipType": "OWNER",
                        "identifier": "urn:li:userGeneratedContent"
                    }
                ]
            }
        }
        response = requests.post(register_url, headers=headers, data=json.dumps(data))
        if response.status_code == 200:
            upload_url = response.json()['value']['uploadMechanism']['com.linkedin.digitalmedia.uploading.MediaUploadHttpRequest']['uploadUrl']
            asset = response.json()['value']['asset']
            return upload_url, asset
        raise Exception(f"Failed to register media: {response.text}")

    def upload_images(self, records_data,logo_url,advertisement_file_path):
        image_files = self.generate_images(records_data,logo_url,advertisement_file_path)
        if not image_files:
            print("No images generated. Exiting upload process.")
            return []
        media_assets = []
        for image_file in image_files:
            upload_url, asset = self.register_media(image_file)
            if upload_url and asset:
                with open(image_file, "rb") as file:
                    headers = {
                        'Authorization': f'Bearer {self.access_token}',
                        'Content-Type': 'application/octet-stream'
                    }
                    response = requests.put(upload_url, headers=headers, data=file)
                if response.status_code == 201:
                    print(f"Uploaded {image_file} successfully!")
                    media_assets.append(asset)
                else:
                    print(f"Failed to upload {image_file}: {response.text}")
            else:
                print(f"Skipping upload for {image_file} due to failed registration.")
        return media_assets

    def share_post(self, media_assets):
        if not media_assets:
            print("No media assets to post. Exiting.")
            return
        url = "https://api.linkedin.com/v2/ugcPosts"
        headers = self.headers.copy()
        headers['Content-Type'] = 'application/json'
        payload = {
            "author": f"urn:li:group:{group_id}",
            "lifecycleState": "PUBLISHED",
            "specificContent": {
                "com.linkedin.ugc.ShareContent": {
                    "shareCommentary": {
                        "text": self.description
                    },
                    "shareMediaCategory": "IMAGE",
                    "media": [
                        {"status": "READY", "media": asset} for asset in media_assets
                    ]
                }
            },
            "visibility": {
                "com.linkedin.ugc.MemberNetworkVisibility": "PUBLIC"
            }
        }
        response = requests.post(url, headers=headers, data=json.dumps(payload))
        if response.status_code == 201:
            print("Post shared successfully!")
        else:
            print(f"Failed to share post: {response.text}")

    def get_user_id(self):
        url = "https://api.linkedin.com/v2/userinfo"
        response = requests.get(url, headers=self.headers)
        if response.status_code == 200:
            self.user_id = response.json().get("sub")
        else:
            print(f"Failed to retrieve LinkedIn user ID: {response.text}")

    def main_func(self, records_data,logo_url,advertisement_file_path,group_id):
        self.get_user_id()
        if self.user_id:
            media_assets = self.upload_images(records_data,logo_url,advertisement_file_path)
            self.share_post(media_assets,group_id)
        else:
            print("Failed to retrieve LinkedIn user ID.")
```

```python
def generate_advertisement_image(logo_url, text,font_url):
    # Update the text message
    para = textwrap.wrap(text, width=40)  # Adjust width for better line management

    # Increase canvas size
    width, height = 1080, 1080
    im = Image.new('RGB', (width, height), color='#003C57')
    draw = ImageDraw.Draw(im)

    # Load and resize the logo from the provided URL
    response = requests.get(logo_url)
    if response.status_code == 200:
        try:
            logo = Image.open(BytesIO(response.content))
            logo = logo.resize((250, 250))  # Resize logo to 250x250 pixels
            im.paste(logo, ((width - logo.width) // 2, 10))  # Center the logo at the top
        except Exception as e:
            print("Une erreur est survenue lors du traitement de l'image:", e)
    else:
        print("La requête HTTP a échoué avec le code:", response.status_code)

    # Utiliser une taille de police plus grande pour une meilleure visibilité
    title_font = load_font_from_url(font_url, font_size=35)  # Bold font for title
    text_font = load_font_from_url(font_url, font_size=35)   # Bold font for text
    watermark_font = load_font_from_url(font_url, font_size=18)  # Regular font for watermark

    # Handle title text wrapping and drawing
    current_h, pad = 110+logo.height, 5  # Start below the logo
    for line in para:
        # Utilisez textbbox pour obtenir les dimensions de la boîte englobante
        bbox = draw.textbbox((0, current_h), line, font=title_font)
        w, h = bbox[2] - bbox[0], bbox[3] - bbox[1]
        draw.text(((width - w) // 2, current_h), line, fill='#EBB016', font=title_font)
        current_h += h + pad  # Increment current height position

    # Add watermark in the bottom-right corner
    watermark_text = "Source: TetrisNews"
    bbox = draw.textbbox((width, height), watermark_text, font=watermark_font)
    w, h = bbox[2] - bbox[0], bbox[3] - bbox[1]
    draw.text((width - w - 10, height - h - 10), watermark_text, fill='white', font=watermark_font)

    # Return the generated image
    return im


```
